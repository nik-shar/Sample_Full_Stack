```python
from fetch_data import fetch_wind_forecasts, fetch_fuelhh_actuals
from clean_data import clean_forecasts, clean_actuals
from helpers import build_full_dataset

# 1. Fetch
raw_forecasts = fetch_wind_forecasts(
    publish_from="2025-01-01T00:00:00Z",
    publish_to="2025-01-07T00:00:00Z",  # start small (important)
)

raw_actuals = fetch_fuelhh_actuals(
    publish_from="2025-01-01T00:00:00Z",
    publish_to="2025-01-07T00:00:00Z",
    settlement_from="2025-01-01",
    settlement_to="2025-01-07",
    fuel_type="WIND",
)

# 2. Clean
forecasts = clean_forecasts(raw_forecasts)
actuals = clean_actuals(raw_actuals)

# 3. Join
joined = build_full_dataset(actuals, forecasts)

print("Done. Shape:", joined.shape)
```

```python
import os

os.makedirs("analysis/data", exist_ok=True)

joined.to_parquet("analysis/data/joined.parquet", index=False)
forecasts.to_parquet("analysis/data/forecasts_clean.parquet", index=False)
actuals.to_parquet("analysis/data/actuals_clean.parquet", index=False)
```
